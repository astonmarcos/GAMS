Produto interno bruto (PIB) dividido pelo número total de pessoas empregadas na economia. Esses dados são ajustados pela inflação até o ano de 2021

fonte: https://ourworldindata.org/grapher/gdp-per-person-employed-constant-ppp

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/gdp-per-person-employed-constant-ppp.csv")
```

```python
filtro_us = df.loc[df['Entity'] == 'United States']
```

```python
print(filtro_us.head())
```

```python
filtro_us.plot(x='Year', y='GDP per person employed (constant 2017 PPP $)', kind='line', legend=False)
plt.title('Produto interno bruto (PIB) dividido pelo número total de pessoas empregadas na economia.')
plt.show()
```
